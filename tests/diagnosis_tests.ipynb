```python
import numpy as np
import pandas as pd
import sys
import os.path

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../disease_diagnosis/diagnosis_system/dataset' + os.path.sep + 'dataset.csv')

cols = df.columns
data = df[cols].to_numpy().flatten()

s = pd.Series(data)
s = s.str.strip()
s = s.values.reshape(df.shape)

df = pd.DataFrame(s, columns=df.columns)
df.head()
```

```python
df = df.fillna(0)
df.head()
```

```python
df_s = pd.read_csv('../disease_diagnosis/diagnosis_system/dataset/Symptom-severity.csv')
df_s.head()
```

```python
symptom_list = list(df_s['Symptom'].unique())
symptom_choices = []
for i in range(len(symptom_list)):
    symptom_choices.append((i+1, symptom_list[i]))
    
list(symptom_choices)
```

```python
symptoms = df_s['Symptom']

for i in range(len(symptoms)):
    weight = df_s['weight'][i]
    df = df.replace(symptoms[i], weight)
df.head()
```

```python
diseases = df['Disease'].unique()
list(diseases)
```
